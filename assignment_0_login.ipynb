```python
import re

# Registartion function
    
def registration():
    temp=0
    
    pattern = '^[a-z]+[\._]?[a-z 0-9]+[@]\w+[.]\w{2,3}$'
    

    # Registering the user with email id and password

    # validating user email id


    username=input('Enter Email :')

    with open('file.txt','r') as f2:

        if re.search(pattern, username):

            for line in f2:

                if username in line:

                    print("Username already Available")
                    temp=1
                    f2.close()
                    break
        else:
            print('Enter valid email address')
            temp=1
        

    # validating user password
    
    specials = ['@','!','#','^','%','$','&','*','(',')']

    if temp != 1:
        password=input('Enter password :')
        if len(password) <16 and len(password) >5:

            if re.search('[0-9]',password):

                if re.search('[A-Z]',password):

                    if re.search('[a-z]',password):

                        if any(char in specials for char in password):

                            with open('file.txt', 'a') as f:

                                f.write(f"\n{username},{password}")
 
                                print("Registration is Successful")
                                
                        else:
                            print("Password should have special letter")
                    else:
                        print("Password should have lower case letter")

                else:
                    print("Password should have capital letter")    
            else:
                print("Password should have Number")

        else:
            print('password length should be between 8 to 16')

        
# forgot password function

def forgot_password():
    
    forgot_mail=input('Enter the registered mail id:\n')

    with open('file.txt','r') as f3:

        for line in f3:

            if forgot_mail in line:
                pas=line.split(',')

                print("The Registered password is",pas[-1])
                f3.close()
                break
        else:
            print("Email Id is not valid")


# login function

def login():
    login_mail=input('Login E-mail:')
    login_pass=input('Login Password:')

    with open('file.txt','r') as f1:
        
        for line in f1:
            
            if login_mail and login_pass in line:
                
                # removing escape sequence from line
                escape = re.sub(r"(\\n|\\r|\\t|\\)", "", line).strip()             
                check=escape.split(',')
                
                if login_mail ==check[0] and login_pass == check[-1]:
                    print("login Sucessful")
                    break

        else:
                new_input=int(input("Incorrect Credentials ,select the following : \n 1.Registration \n 2.Forgot Password \n"))

                if new_input == 1:
                    registration()
                elif new_input == 2:
                    forgot_password()

                else:
                    print("kindly register to continue !")
                    
# Main Program

user_input=int(input("Enter your action \n 1.Registration \n 2.Login \n 3.Forget Password\n"))
new_input =0
if user_input == 1:
    registration()
elif user_input == 2:
    login()
elif user_input == 3:
    forgot_password()
else:
    print("Enter correct input !")
```
